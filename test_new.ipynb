```python
import pandas as pd
import numpy as np
import scipy as sp
from tkinter import *
import tkinter as tk
from tkinter import messagebox
from tkinter import filedialog
import matplotlib.pyplot as plt

def menu():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    label=Label(win1)
    label=Label(win1,text='主菜單',bg='yellow',width=30,font='Helvetica 30 bold')
    label.grid(column=5,row=2)
    btn1=Button(win1,text='開始',width=25,command=msgshow)
    btn2=Button(win1,text='重新選擇檔案',width=25,command=load)
    btn3=Button(win1,text='離開',width=25,command=win1.destroy)
    btn1.grid(column=5,row=15)
    btn2.grid(column=5,row=16)
    btn3.grid(column=5,row=17)
    win1.mainloop()

def menu1():
    global win1
    win1.destroy()
    global data
    data=pd.read_csv(file_path)
    data=pd.DataFrame(data)
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    label=Label(win1)
    label=Label(win1,text='主菜單',bg='yellow',width=30,font='Helvetica 30 bold')
    label.grid(column=5,row=2)
    btn1=Button(win1,text='開始',width=25,command=msgshow)
    btn2=Button(win1,text='重新選擇檔案',width=25,command=load)
    btn3=Button(win1,text='離開',width=25,command=win1.destroy)
    btn1.grid(column=5,row=15)
    btn2.grid(column=5,row=16)
    btn3.grid(column=5,row=17)
    win1.mainloop()
    
def msgshow():
    global win1
    win1.destroy()
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label=Label(win1,text='第一步、請選擇處理缺失資料的方法',bg='yellow',width=80,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='完整個體法',width=20,command=listwise)
    btn2=Button(win1,text='配對變數刪除',width=20,command=pairwise_select)
    btn3=Button(win1,text='單一插補法',width=20,command=singleimputer_select)
    btn4=Button(win1,text='多重插補法',width=20,command=mice_select)
    btn5=Button(win1,text='最大期望演算法',width=20,command=em)
    btn6=Button(win1,text='回到主菜單(重新選擇方法)',width=20,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    btn6.grid(column=0,row=9)
    win1.mainloop()
def msgshow1():
    global win1
    win1.destroy()
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label=Label(win1,text='接著進行第一步，對其餘變數選擇處理缺失資料的方法',bg='yellow',width=80,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='完整個體法',width=20,command=listwise)
    btn2=Button(win1,text='配對變數刪除',width=20,command=pairwise_select)
    btn3=Button(win1,text='單一插補法',width=20,command=singleimputer_select)
    btn4=Button(win1,text='多重插補法',width=20,command=mice_select)
    btn5=Button(win1,text='最大期望演算法',width=20,command=em)
    btn6=Button(win1,text='回到主菜單(重新選擇方法)',width=20,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    btn6.grid(column=0,row=9)
    win1.mainloop()
def login():
    if ent1.get()=="NCHU" and ent2.get()=="user":
        tk.messagebox.showinfo(title='welcome',message='登入成功')
        start()
    else:
        tk.messagebox.showerror(title='Error',message='帳號或密碼錯誤!')
    
    
def listwise():
    global data
    data.dropna(inplace=True)
    da1_show()
    
def da1_show():
    global win1
    global data
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label1=Label(win1,text='以下為您的資料',bg='yellow',width=20,font='Helvetica 10 bold')
    label2=Label(win1,text=data,bg='white',width=100,font='Helvetica 10 bold')
    btn1=Button(win1,text='繼續',width=20,command=transform)
    label1.grid(column=0,row=1)
    label2.grid(column=0,row=2)
    btn1.grid(column=0,row=3)
    win1.mainloop()
def da2_show():
    global win1
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label1=Label(win1,text='以下為您的資料',bg='yellow',width=20,font='Helvetica 10 bold')
    label2=Label(win1,text=data,bg='white',width=100,font='Helvetica 10 bold')
    btn1=Button(win1,text='繼續',width=20,command=detect)
    label1.grid(column=0,row=1)
    label2.grid(column=0,row=2)
    btn1.grid(column=0,row=3)
    win1.mainloop()    
def pairwise_select():
    global win1
    win1.destroy()
    global data
    global data_columns_name
    global var_name
    data_columns_name=list(data.columns)    
    var_name=list(np.arange(len(data_columns_name)))
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label=Label(win1,text='請選擇欲分析的變數',bg='yellow',width=50,font='Helvetica 10 bold')
    label.pack(anchor='w')
    for i in range(len(data_columns_name)):
        var_name[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name[i])
        cb.pack(anchor='w')
    btn=Button(win1,text='完成',width=20,command=check)
    btn.pack()
    win1.mainloop()

def check():
    global choose_column
    choose_column=[]
    global data
    global data_columns_name
    global var_name
    for i in range(len(data_columns_name)):
        if var_name[i].get()==1:
            choose_column.append(i)
    data=data.iloc[:,choose_column]
    listwise()
    
def singleimputer_select():
    global win1
    win1.destroy()
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label=Label(win1,text='請選擇欲插捕之值',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='mean',width=20,command=mean)
    btn2=Button(win1,text='median',width=20,command=median)
    btn3=Button(win1,text='most_frequent',width=20,command=most_frequent)
    btn4=Button(win1,text='constant',width=20,command=constant)
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=20,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    win1.mainloop()
def mean():
    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
    global data
    global data_columns_name
    data_columns_name=list(data.columns)
    imputer.fit(data)
    data=imputer.transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()
def median():
    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(missing_values=np.nan,strategy='median')
    global data
    global data_columns_name
    data_columns_name=list(data.columns)
    imputer.fit(data)
    data=imputer.transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()
def most_frequent():
    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
    global data
    global data_columns_name
    data_columns_name=list(data.columns)
    imputer.fit(data)
    data=imputer.transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)  
    da1_show()
def constant():
    from sklearn.impute import SimpleImputer
    imputer = SimpleImputer(missing_values=np.nan,strategy='constant')
    global data
    global data_columns_name
    data_columns_name=list(data.columns)
    imputer.fit(data)
    data=imputer.transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()
def em():
    global data
    import impyute as impy
    global data_columns_name
    data_columns_name=list(data.columns)
    data=np.array(data,dtype='float')
    data=impy.em(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()
def mice_select():
    global win1
    win1.destroy()
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label1=Label(win1,text='請選擇想迭代幾次(預設10次)',bg='yellow',width=50,font='Helvetica 10 bold')
    label1.grid(column=0,row=2)
    btn1=Button(win1,text='10次',width=20,command=mice)
    btn2=Button(win1,text='20次',width=20,command=mice2)
    btn3=Button(win1,text='50次',width=20,command=mice3)
    btn4=Button(win1,text='100次',width=20,command=mice4)
    btn5=Button(win1,text='500次',width=20,command=mice5)
    btn6=Button(win1,text='1000次',width=20,command=mice6)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    btn6.grid(column=0,row=9)
    win1.mainloop()
def mice():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=10, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()
def mice2():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=20, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()    
def mice3():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=50, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show() 
def mice4():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=100, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()  
def mice5():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=500, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()  
def mice6():
    global data
    import numpy as np
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer
    from sklearn.linear_model import LinearRegression
    global data_columns_name
    data_columns_name=list(data.columns)
    lr = LinearRegression()
    imp = IterativeImputer(estimator=lr,missing_values=np.nan, max_iter=1000, verbose=2, imputation_order='roman',random_state=0)
    data=imp.fit_transform(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da1_show()  
def load():
    global win1
    global data
    global file_path
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    file_path=filedialog.askopenfilename()
    data=pd.read_csv(file_path)
    data=pd.DataFrame(data)
    label=Label(win1)
    label=Label(win1,text='請問此資料是否含有具時間序列性質之變數?',bg='yellow',width=100,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='是',width=25,command=detect2)   #這裡ts_select先改掉detect2
    btn2=Button(win1,text='否',width=25,command=data_show2)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    win1.mainloop()
def ts_select():
    global win1
    global data
    global data_columns_name
    global var_name
    win1.destroy()
    data_columns_name=list(data.columns)
    var_name=list(np.arange(len(data_columns_name)))
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    label=Label(win1,text='請勾選具時間序列性質之變數(將先對您選擇之變數進行近值插補)',bg='yellow',width=100,font='Helvetica 10 bold')
    label.pack(anchor='w')
    for i in range(len(data_columns_name)):
        var_name[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name[i])
        cb.pack(anchor='w')
    btn=Button(win1,text='完成',width=20,command=ts)
    btn.pack()
    win1.mainloop()
def ts():
    global choose_column
    choose_column=[]
    global data
    global data_columns_name
    global var_name
    for i in range(len(data_columns_name)):
        if var_name[i].get()==1:
            choose_column.append(i)
    data.iloc[:,choose_column]=data.iloc[:,choose_column].fillna(method='ffill')
    data_show()
def data_show():
    global win1
    global data_columns_name
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    label=Label(win1)
    label=Label(win1,text='以下為處理完具時間序列性變數後您的資料詳情',bg='yellow',width=100,font='Helvetica 10 bold')
    label2=Label(win1,text=data.describe(),bg='white',width=100,font='Helvetica 10 bold')
    btn=Button(win1,text='繼續',width=15,command=msgshow1)
    label.pack()
    label2.pack()
    btn.pack()
    win1.mainloop()
def data_show2():
    global win1
    global data_columns_name
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    label=Label(win1)
    label=Label(win1,text='以下為您選取的資料詳情',bg='yellow',width=100,font='Helvetica 10 bold')
    label2=Label(win1,text=data.describe(),bg='white',width=100,font='Helvetica 10 bold')
    btn=Button(win1,text='繼續',width=15,command=menu)
    label.pack()
    label2.pack()
    btn.pack()
def start(): 
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='歡迎使用此介面，注意請選擇匯入資料全為數值型之CSV檔',bg='yellow',width=100,font='Helvetica 10 bold')
    label.pack()
    btn=Button(win1,text='選擇檔案',width=15,command=load)
    btn.pack()
    win1.mainloop()

def transform():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第二步、請選擇想使用的資料分佈轉換',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='Z-score Normalization',width=25,command=zscore)
    btn2=Button(win1,text='Mean Normalization',width=25,command=mean_normalization)
    btn3=Button(win1,text='Min-Max Normalization',width=25,command=min_max)
    btn4=Button(win1,text='Unit Vector Normalization',width=25,command=unit_vector)
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    win1.mainloop()
def zscore():
    global win1
    win1.destroy()
    global data
    import scipy.stats as stats
    data = stats.zscore(data)
    da2_show()
def mean_normalization():
    global win1
    win1.destroy()
    global data
    from sklearn import preprocessing
    global data_columns_name
    data_columns_name=list(data.columns)
    data=preprocessing.normalize(data)
    data=pd.DataFrame(data,columns=data_columns_name)
    da2_show()
    
def min_max():
    global win1
    win1.destroy()
    global data
    from sklearn.preprocessing import MinMaxScaler
    min_max = MinMaxScaler()
    data=pd.DataFrame(min_max.fit_transform(data),columns=data.keys())
    da2_show()
    
def unit_vector():
    global win1
    win1.destroy()
    global data
    import numpy as np
    global data_columns_name
    data_columns_name=list(data.columns)
    data=np.array(data,dtype='float')
    y=np.empty(len(data[0,:]), dtype=None) 
    for i in range(0,len(data[0,:])):
        y[i] = np.linalg.norm(data[:,i])
    for i in range(0,len(data[0,:])):
        data[:,i]=data[:,i]/y[i]
    data=pd.DataFrame(data,columns=data_columns_name)
    da2_show()
    
def detect():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第三步、請選擇想使用的異常資料檢測方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='盒狀圖法',width=25,command=plot_page) # 先把這個改掉plot_page
    btn2=Button(win1,text='Mean & Standard Deviation',width=25,command=msd)
    btn3=Button(win1,text='馬氏距離法',width=25,command=most_frequent)
    btn4=Button(win1,text='DBSCAN',width=25,command=constant)
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    win1.mainloop()
def plot_page():
    global win1
    global data
    global choose_column
    global var_name
    global data_columns_name
    data_columns_name=list(data.columns)
    var_name=list(np.arange(len(data_columns_name)))
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請點選變數查看箱型圖(照順序排列)',bg='yellow',width=50,font='Helvetica 10 bold')
    label2=Label(win1,text='(箱型圖之極端值為超出第一四分位或第三四分位之1.5IQR之值)',bg='yellow',width=100,font='Helvetica 10 bold')
    label.pack(anchor='center')
    label2.pack(anchor='center')
    for i in range(len(data_columns_name)):
        var_name[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name[i])
        cb.pack(anchor='w')
    btn1=Button(win1,text='勾選完成並觀看盒狀圖',width=20,command=plot)
    btn2=Button(win1,text='繼續下個步驟',width=20,command=box_select)
    btn3=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.pack(anchor='center')
    btn2.pack(anchor='center')
    btn3.pack(anchor='center')
    win1.mainloop()
def plot():
    import matplotlib.pyplot as plt
    from matplotlib.figure import Figure 
    from matplotlib.backends.backend_tkagg import (FigureCanvasTkAgg,NavigationToolbar2Tk) 
    global win1
    global data
    global var_name
    global data_columns_name
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")
    choose_column=[]
    for i in range(len(data_columns_name)):
        if var_name[i].get()==1:
            choose_column.append(i)
    fig = plt.figure(figsize = (5,5), dpi = 100) 
    plt.boxplot(data.iloc[:,choose_column])
    plt.title("Box plot")
    canvas = FigureCanvasTkAgg(fig,master=win1)
    canvas.draw()
    canvas.get_tk_widget().pack()
    btn=Button(win1,text="返回",width=25,command=plot_page)
    btn.pack(anchor='center')
    win1.mainloop()
    
def box_select():
    global win1
    global ent1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入想剔除幾個IQR外之極端值?(請輸入數字，預設為1.5)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    btn1=Button(win1,text='確定並繼續',width=15,command=box_result)
    btn2=Button(win1,text='不刪除極端值並繼續(直接點選此按鈕勿於上方輸入數字)',width=50,command=box_result)
    btn3=Button(win1,text='返回觀看盒狀圖',width=15,command=plot_page)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    win1.mainloop()
def box_result():
    global data_columns_name
    global ent1
    global data
    data_columns_name=list(data.columns)
    for i in range(len(data_columns_name)):
        mini=np.percentile(data.iloc[:,i],25)-float(ent1.get())*(np.percentile(data.iloc[:,i],75)-np.percentile(data.iloc[:,i],25))
        maxi=np.percentile(data.iloc[:,i],75)+float(ent1.get())*(np.percentile(data.iloc[:,i],75)-np.percentile(data.iloc[:,i],25))
        ucl=data.iloc[:,i]>maxi
        lcl=data.iloc[:,i]<mini
        data.iloc[:,i][ucl|lcl]=np.nan
    data.dropna(inplace=True)
    data=pd.DataFrame(data,columns=data_columns_name) 
    detect2()
def msd():
    global data_columns_name
    global ent1
    global data
    data_columns_name=list(data.columns) 
    for i in range(len(data_columns_name)):
        mini=np.percentile(data.iloc[:,i],25)-3*(np.percentile(data.iloc[:,i],75)-np.percentile(data.iloc[:,i],25))
        maxi=np.percentile(data.iloc[:,i],75)+3*(np.percentile(data.iloc[:,i],75)-np.percentile(data.iloc[:,i],25))
        ucl=data.iloc[:,i]>maxi
        lcl=data.iloc[:,i]<mini
        data.iloc[:,i][ucl|lcl]=np.nan
    data.dropna(inplace=True)
    data=pd.DataFrame(data,columns=data_columns_name) 
    detect2()

def mahalanobis_():
    global ent1
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入超過多少mahalanobis值刪除(輸入數字)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    btn1=Button(win1,text='確定',width=15,command=mahalanobis_result)
    btn1.grid(column=0,row=4)
    win1.mainloop()


def Mahalanobis(X, meanCol, incov):
    from scipy.spatial.distance import mahalanobis
    m = []
    for i in range(X.shape[0]):
        m.append(mahalanobis(X.iloc[i,:], meanCol, incov)**2)
    return(m)
def mahalanobis_result():
    global data
    global ent1
    x = data
    x = x.iloc[:,1:]
    Sx = x.cov().values
    Sx = sp.linalg.inv(Sx)
    mean = x.mean().values
    mR = Mahalanobis(x,mean,Sx)
    data['Mahalanobis'] = mR
    data = data[ data['Mahalanobis'] < float(ent1.get()) ] 
    data = data.drop('Mahalanobis', axis=1)
    detect2()
    

def DBSCAN_():
    global win1
    win1.destroy()
    global ent1
    global ent2
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入eps(輸入數字，預設0.5)',bg='yellow',width=30,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    label=Label(win1,text='請輸入min_samples值(輸入數字，預設10)',bg='yellow',width=40,font='Helvetica 10 bold')
    label.grid(column=0,row=4)
    ent2=Entry(win1)
    ent2.grid(column=0,row=5)
    btn1=Button(win1,text='確定',width=15,command=DBSCAN_result)
    btn1.grid(column=0,row=6)
    win1.mainloop()
    
def DBSCAN_result():
    global data
    global ent1
    global ent2
    from collections import Counter
    from sklearn.cluster import DBSCAN
    model = DBSCAN(eps=float(ent1.get()), min_samples=int(ent2.get())).fit(data) 
    print (Counter(model.labels_))     
    data = data[model.labels_!=-1]     #dataframe
    detect2()

def detect2():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第四步、請選擇想使用的相依檢測方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='統計迴歸法',width=25,command=stat_regression_method)
    btn2=Button(win1,text='多變量分析法',width=25,command=stat_multi_method)
    btn3=Button(win1,text='時間序列分析法',width=25,command=time_series_method) #time_series_method
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn5.grid(column=0,row=7)
    win1.mainloop()

#第五步驟:回歸
def stat_regression_method():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第五步、請選擇想使用的迴歸方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='多重線性迴歸法',width=25,command=regression_method_1)
    btn2=Button(win1,text='多項式迴歸法',width=25,command=regression_method_1)
    btn3=Button(win1,text='逐步迴歸法',width=25,command=regression_method_13)
    btn4=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    win1.mainloop()

###########################多重線性迴歸法
def regression_method_1():
    global win1
    global data
    global choose_column_Y
    global var_name_Y
    global data_columns_name
    data_columns_name=list(data.columns)
    var_name_Y=list(np.arange(len(data_columns_name)))
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請先點選一個 Y 變數(即為反應變數，想知道的結果項)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.pack(anchor='center')
    for i in range(len(data_columns_name)):
        var_name_Y[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name_Y[i])
        cb.pack(anchor='w')
    btn1=Button(win1,text='繼續下個步驟',width=20,command=regression_method_11)
    btn2=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.pack(anchor='center')
    btn2.pack(anchor='center')
    win1.mainloop()

def regression_method_11():
    global win1
    global data
    global choose_column_x
    global var_name
    global data_columns_name
    data_columns_name=list(data.columns)
    var_name=list(np.arange(len(data_columns_name)))
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請點選一個或多個 X 變數(即為自變項)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.pack(anchor='center')
    for i in range(len(data_columns_name)):
        var_name[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name[i])
        cb.pack(anchor='w')
    btn1=Button(win1,text='繼續下個步驟',width=20,command=linear_regression)    #回歸方法把 X 跟 Y 寫進函數
    btn2=Button(win1,text='返回上一個步驟(重新挑選 Y 變數)',width=30,command=regression_method_1)
    btn3=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.pack(anchor='center')
    btn2.pack(anchor='center')
    btn3.pack(anchor='center')
    win1.mainloop()

def linear_regression():
    import statsmodels.api as sm
    global win1
    global data
    global var_name
    global var_name_Y
    global data_columns_name
    global choose_column_Y
    global choose_column_x
    global Y_select
    global x_select
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")

    choose_column_Y=[]
    for i in range(len(data_columns_name)):
        if var_name_Y[i].get()==1:
            choose_column_Y.append(i)
    Y_select = data.iloc[:,choose_column_Y]
    #print(Y_select)

    choose_column_x=[]
    for j in range(len(data_columns_name)):
        if var_name[j].get()==1:
            choose_column_x.append(j)  
          
    x_select = data.iloc[:,choose_column_x]
    #print(x_select)
    reg = sm.OLS(Y_select,x_select)
    model = reg.fit()
    print(model.summary())

    btn1=Button(win1,text='返回上一個步驟(重新挑選 Y 變數)',width=30,command=regression_method_1)
    btn1.pack(anchor='center')
    win1.mainloop()
###########################多重線性迴歸法 到這裡 Ending

###########################逐步回歸
#import statsmodels.formula.api as sm
def regression_method_13():
    global win1
    global data
    global choose_column_Y
    global var_name_Y
    global data_columns_name
    data_columns_name=list(data.columns)
    var_name_Y=list(np.arange(len(data_columns_name)))
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請先點選一個 Y 變數，其他則為X變數',bg='yellow',width=50,font='Helvetica 10 bold')
    label.pack(anchor='center')
    for i in range(len(data_columns_name)):
        var_name_Y[i]=tk.IntVar()
        cb=Checkbutton(win1,text=data_columns_name[i],variable=var_name_Y[i])
        cb.pack(anchor='w')
    btn1=Button(win1,text='繼續下個步驟',width=20,command=stepwise_regression)    
    btn2=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.pack(anchor='center')
    btn2.pack(anchor='center')
    win1.mainloop()

def stepwise_regression():
    import statsmodels.api as sm
    global win1
    global data
    global var_name_Y
    global data_columns_name
    global choose_column_Y
    global y
    global X
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("800x800")

    choose_column_Y=[]
    for i in range(len(data_columns_name)):
        if var_name_Y[i].get()==1:
            choose_column_Y.append(i)
    y = data.iloc[:,choose_column_Y]
    X = data.drop(data.iloc[:,choose_column_Y], axis=1)
    print(" Y 變 數\n",y)
    print("===================================================")
    print(" X 變 數\n",X)
    print("===================================================")

    btn1=Button(win1,text='向前選取法 (Forward Selection)',width=28,command=forwardSelection_go)    
    btn2=Button(win1,text='向後選取法 (backward Selection)',width=28,command=backwardSelection_go)
    btn3=Button(win1,text='返回上一個步驟 (重新挑選 Y 變數)',width=30,command=regression_method_13)
    btn4=Button(win1,text='回到主菜單 (重新選擇方法)',width=25,command=menu1)
    btn1.pack(anchor='center')
    btn2.pack(anchor='center')
    btn3.pack(anchor='center')
    btn4.pack(anchor='center')
    win1.mainloop()

def __varcharProcessing__(X,varchar_process= "dummy_dropfirst"):
    varchar_process = "dummy_dropfirst"
    dtypes = X.dtypes
    if varchar_process == "drop":   
        X = X.drop(columns = dtypes[dtypes == np.object].index.tolist())
        print("Character Variables (Dropped):", dtypes[dtypes == np.object].index.tolist())
    elif varchar_process == "dummy":
        X = pd.get_dummies(X,drop_first=False)
        print("Character Variables (Dummies Generated):", dtypes[dtypes == np.object].index.tolist())
    elif varchar_process == "dummy_dropfirst":
        X = pd.get_dummies(X,drop_first=True)
        print("Character Variables (Dummies Generated, First Dummies Dropped):", dtypes[dtypes == np.object].index.tolist())
    else: 
        X = pd.get_dummies(X,drop_first=True)
        print("Character Variables (Dummies Generated, First Dummies Dropped):", dtypes[dtypes == np.object].index.tolist())
    
    X["intercept"] = 1
    cols = X.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    X = X[cols]
    return X

def __forwardSelectionRaw__(X, y, model_type ="linear",elimination_criteria = "aic", sl=0.05):
    import statsmodels.api as sm
    
    iterations_log = ""
    cols = X.columns.tolist()
    
    def regressor(y,X, model_type=model_type):

        if model_type == "linear":
            regressor = sm.OLS(y, X).fit()
        elif model_type == "logistic":
            regressor = sm.Logit(y, X).fit()
        else:
            print("\nWrong Model Type : "+ model_type +"\nLinear model type is seleted.")
            model_type = "linear"
            regressor = sm.OLS(y, X).fit()
        return regressor
    
    selected_cols = ["intercept"]
    other_cols = cols.copy()
    other_cols.remove("intercept")
    
    model = regressor(y, X[selected_cols])
    
    if elimination_criteria == "aic":
        criteria = model.aic
    elif elimination_criteria == "bic":
        criteria = model.bic
    elif elimination_criteria == "r2" and model_type =="linear":
        criteria = model.rsquared
    elif elimination_criteria == "adjr2" and model_type =="linear":
        criteria = model.rsquared_adj
    
    
    for i in range(X.shape[1]):
        pvals = pd.DataFrame(columns = ["Cols","Pval"])
        for j in other_cols:
            model = regressor(y, X[selected_cols+[j]])
            pvals = pvals.append(pd.DataFrame([[j, model.pvalues[j]]],columns = ["Cols","Pval"]),ignore_index=True)
        pvals = pvals.sort_values(by = ["Pval"]).reset_index(drop=True)
        pvals = pvals[pvals.Pval<=sl]
        if pvals.shape[0] > 0:
            
            model = regressor(y, X[selected_cols+[pvals["Cols"][0]]])
            iterations_log += str("\nEntered : "+pvals["Cols"][0] + "\n")    
            iterations_log += "\n\n"+str(model.summary())+"\nAIC: "+ str(model.aic) + "\nBIC: "+ str(model.bic)+"\n\n"
                    
        
            if  elimination_criteria == "aic":
                new_criteria = model.aic
                if new_criteria < criteria:
                    print("Entered :", pvals["Cols"][0], "\tAIC :", model.aic)
                    selected_cols.append(pvals["Cols"][0])
                    other_cols.remove(pvals["Cols"][0])
                    criteria = new_criteria
                else:
                    print("break : Criteria")
                    break
            elif  elimination_criteria == "bic":
                new_criteria = model.bic
                if new_criteria < criteria:
                    print("Entered :", pvals["Cols"][0], "\tBIC :", model.bic)
                    selected_cols.append(pvals["Cols"][0])
                    other_cols.remove(pvals["Cols"][0])
                    criteria = new_criteria
                else:
                    print("break : Criteria")
                    break        
            elif  elimination_criteria == "r2" and model_type =="linear":
                new_criteria = model.rsquared
                if new_criteria > criteria:
                    print("Entered :", pvals["Cols"][0], "\tR2 :", model.rsquared)
                    selected_cols.append(pvals["Cols"][0])
                    other_cols.remove(pvals["Cols"][0])
                    criteria = new_criteria
                else:
                    print("break : Criteria")
                    break           
            elif  elimination_criteria == "adjr2" and model_type =="linear":
                new_criteria = model.rsquared_adj
                if new_criteria > criteria:
                    print("Entered :", pvals["Cols"][0], "\tAdjR2 :", model.rsquared_adj)
                    selected_cols.append(pvals["Cols"][0])
                    other_cols.remove(pvals["Cols"][0])
                    criteria = new_criteria
                else:
                    print("Break : Criteria")
                    break
            else:
                print("Entered :", pvals["Cols"][0])
                selected_cols.append(pvals["Cols"][0])
                other_cols.remove(pvals["Cols"][0])            
                
        else:
            print("Break : Significance Level")
            break
        
    model = regressor(y, X[selected_cols])
    if elimination_criteria == "aic":
        criteria = model.aic
    elif elimination_criteria == "bic":
        criteria = model.bic
    elif elimination_criteria == "r2" and model_type =="linear":
        criteria = model.rsquared
    elif elimination_criteria == "adjr2" and model_type =="linear":
        criteria = model.rsquared_adj
    
    print(model.summary())
    print("AIC: "+str(model.aic))
    print("BIC: "+str(model.bic))
    print("Final Variables:", selected_cols)

    return selected_cols, iterations_log
    
def forwardSelection(X, y, model_type ="linear",elimination_criteria = "aic", varchar_process = "dummy_dropfirst", sl=0.05):
    X = __varcharProcessing__(X,varchar_process)
    return __forwardSelectionRaw__(X, y, model_type = model_type, elimination_criteria = elimination_criteria , sl=sl)

def forwardSelection_go():
   global win1
   global data
   global var_name_Y
   global data_columns_name
   global choose_column_Y
   global y
   global X 
   global final_vars
   global iterations_logs
   
   final_vars, iterations_logs = forwardSelection(X,y, model_type="linear")

def __backwardSelectionRaw__(X, y, model_type ="linear",elimination_criteria = "aic", sl=0.05):
    import statsmodels.api as sm

    iterations_log = ""
    last_eleminated = ""    
    cols = X.columns.tolist()

    def regressor(y,X, model_type=model_type):
        if model_type =="linear":
            regressor = sm.OLS(y, X).fit()
        elif model_type == "logistic":
            regressor = sm.Logit(y, X).fit()
        else:
            print("\nWrong Model Type : "+ model_type +"\nLinear model type is seleted.")
            model_type = "linear"
            regressor = sm.OLS(y, X).fit()
        return regressor
    for i in range(X.shape[1]):
        if i != 0 :          
            if elimination_criteria == "aic":
                criteria = model.aic
                new_model = regressor(y,X)
                new_criteria = new_model.aic
                if criteria < new_criteria:
                    print("Regained : ", last_eleminated)
                    iterations_log += "\n"+str(new_model.summary())+"\nAIC: "+ str(new_model.aic) + "\nBIC: "+ str(new_model.bic)+"\n"
                    iterations_log += str("\n\nRegained : "+last_eleminated + "\n\n")
                    break  
            elif elimination_criteria == "bic":
                criteria = model.bic
                new_model = regressor(y,X)
                new_criteria = new_model.bic
                if criteria < new_criteria:
                    print("Regained : ", last_eleminated)
                    iterations_log += "\n"+str(new_model.summary())+"\nAIC: "+ str(new_model.aic) + "\nBIC: "+ str(new_model.bic)+"\n"
                    iterations_log += str("\n\nRegained : "+last_eleminated + "\n\n")
                    break  
            elif elimination_criteria == "adjr2" and model_type =="linear":
                criteria = model.rsquared_adj
                new_model = regressor(y,X)
                new_criteria = new_model.rsquared_adj
                if criteria > new_criteria:
                    print("Regained : ", last_eleminated)
                    iterations_log += "\n"+str(new_model.summary())+"\nAIC: "+ str(new_model.aic) + "\nBIC: "+ str(new_model.bic)+"\n"
                    iterations_log += str("\n\nRegained : "+last_eleminated + "\n\n")
                    break  
            elif elimination_criteria == "r2" and model_type =="linear":
                criteria = model.rsquared
                new_model = regressor(y,X)
                new_criteria = new_model.rsquared
                if criteria > new_criteria:
                    print("Regained : ", last_eleminated)
                    iterations_log += "\n"+str(new_model.summary())+"\nAIC: "+ str(new_model.aic) + "\nBIC: "+ str(new_model.bic)+"\n"
                    iterations_log += str("\n\nRegained : "+last_eleminated + "\n\n")
                    break   
            else: 
                new_model = regressor(y,X)
            model = new_model
            iterations_log += "\n"+str(model.summary())+"\nAIC: "+ str(model.aic) + "\nBIC: "+ str(model.bic)+"\n"
        else:
            model = regressor(y,X)
            iterations_log += "\n"+str(model.summary())+"\nAIC: "+ str(model.aic) + "\nBIC: "+ str(model.bic)+"\n"
        maxPval = max(model.pvalues)
        cols = X.columns.tolist()
        if maxPval > sl:
            for j in cols:
                if (model.pvalues[j] == maxPval):
                    print("Eliminated :" ,j)
                    iterations_log += str("\n\nEliminated : "+j+ "\n\n")
                    
                    del X[j]
                    last_eleminated = j
        else:
            break
    print(str(model.summary())+"\nAIC: "+ str(model.aic) + "\nBIC: "+ str(model.bic))
    print("Final Variables:", cols)
    iterations_log += "\n"+str(model.summary())+"\nAIC: "+ str(model.aic) + "\nBIC: "+ str(model.bic)+"\n"
    return cols, iterations_log

def backwardSelection(X, y, model_type ="linear",elimination_criteria = "aic", varchar_process = "dummy_dropfirst", sl=0.05):
    X = __varcharProcessing__(X,varchar_process = varchar_process)
    return __backwardSelectionRaw__(X, y, model_type = model_type,elimination_criteria = elimination_criteria , sl=sl)

def backwardSelection_go():
   global win1
   global data
   global var_name_Y
   global data_columns_name
   global choose_column_Y
   global y
   global X 
   global final_vars
   global iterations_logs
   
   final_vars, iterations_logs = backwardSelection(X,y, model_type="linear")
###########################逐步回歸到這裡 Ending

#第五步驟:多變量分析
def stat_multi_method():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第五步、請選擇想使用的多變量分析方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='正規變量分析法()',width=25,command=regression_method_1)
    btn2=Button(win1,text='正規相關性分析法',width=25,command=regression_method_1)
    btn3=Button(win1,text=' 主成分析法PCA',width=25,command=regression_method_1)
    btn4=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    win1.mainloop()

#第五步驟:時間序列
def time_series_method():
    global win1
    win1.destroy()
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='第五步、請選擇想使用的時間序列方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='簡單移動平均法',width=25,command=simple_select)
    btn2=Button(win1,text='加權移動平均法',width=25,command=WMA_select)
    btn3=Button(win1,text='指數平滑法',width=25,command=SEM_select)
    btn4=Button(win1,text='季節趨勢預測法',width=25,command=SD_select)
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=25,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn3.grid(column=0,row=6)
    btn4.grid(column=0,row=7)
    btn5.grid(column=0,row=8)
    win1.mainloop()


def simple_select():
    global win1
    win1.destroy()
    global ent1
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入想移動平滑窗口的值?(輸入數字)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    btn1=Button(win1,text='確定',width=15,command=SMA_result)
    btn1.grid(column=0,row=4)
    win1.mainloop()

# 這個有問題，但測試跑得出來...
def SMA_result():
    global data
    window = 3 #窗口大小爲 3 的情況下的移動平均值
    # trail-rolling average transform
    rolling = data.rolling(window=window)
    rolling_mean = rolling.mean()
    plt.figure(figsize=(15, 7))
    plt.plot(data)
    plt.plot(rolling_mean, 'r')

#加權
def WMA_select():
    global win1
    win1.destroy()
    global ent1
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入想移動窗口數值?(輸入數字)',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    btn1=Button(win1,text='確定',width=15,command=WMA_result)
    btn1.grid(column=0,row=4)
    win1.mainloop()

def WMA(data, n):
    global windows
    weights = np.array(range(1, n+1))
    sum_weights = np.sum(weights)
    res = data.rolling(window=n).apply(lambda x: np.sum(weights*x) / sum_weights, raw=False)
    return res    
def WMA_result():   
    global data
    global ent1
    data['WMA_window'] = WMA(data['births'], int(ent1.get()))   #這裡可能要改成用挑選的變數名稱，births要改掉
    return(data)

def SEM_select():
    global win1
    win1.destroy()
    global ent1
    win1=Tk() #建立根視窗
    win1.title('研究子計畫')
    win1.geometry("500x500")
    label=Label(win1)
    label=Label(win1,text='請輸入smoothing_level(輸入數字)，預設0.1',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    ent1=Entry(win1)
    ent1.grid(column=0,row=3)
    btn1=Button(win1,text='確定',width=15,command=SEM_result)
    btn1.grid(column=0,row=4)
    win1.mainloop()

def SEM_result():
    from statsmodels.tsa.api import  SimpleExpSmoothing
    global data
    #global ent1
    fit = SimpleExpSmoothing(data["births"]).fit()
    print(fit.fittedvalues) #平滑後的值
    print(fit.forecast()) #預測
    

def SD_select():
    global win1
    win1.destroy()
    win1=Tk()
    win1.title('研究子計畫')
    win1.geometry("1000x1000")
    label=Label(win1)
    label=Label(win1,text='請選擇想要計算的序列方法',bg='yellow',width=50,font='Helvetica 10 bold')
    label.grid(column=0,row=2)
    btn1=Button(win1,text='加法序列',width=20,command=SDPlus_result)
    btn2=Button(win1,text='乘法序列',width=20,command=SDMulti_result)
    btn5=Button(win1,text='回到主菜單(重新選擇方法)',width=20,command=menu1)
    btn1.grid(column=0,row=4)
    btn2.grid(column=0,row=5)
    btn5.grid(column=0,row=8)
    win1.mainloop()
    
def SDPlus_result():
    global data
    from statsmodels.tsa.seasonal import seasonal_decompose
    # 加法序列 = Level + Trend + Seasonality + Error
    result_add = seasonal_decompose(data['births'], model='additive', extrapolate_trend='freq',period=None, freq=12)
    print(result_add.trend)
    print(result_add.seasonal)
    print(result_add.resid)
    print(result_add.observed)             
    plt.rcParams.update({'figure.figsize': (20, 10)})
    result_add.plot().suptitle('\nAdditive Decompose')
    plt.show()

def SDMulti_result():
    global data
    from statsmodels.tsa.seasonal import seasonal_decompose
    # 加法序列 = Level + Trend + Seasonality + Error
    result_add = seasonal_decompose(data['births'], model='multiplicative', extrapolate_trend='freq',period=None, freq=12)
    print(result_add.trend)
    print(result_add.seasonal)
    print(result_add.resid)
    print(result_add.observed)             
    plt.rcParams.update({'figure.figsize': (20, 10)})
    result_add.plot().suptitle('\nMultiplicative Decompose')
    plt.show() 





global data_columns_name
global var_name
win1=Tk() #建立根視窗
win1.title('研究子計畫')
win1.geometry("500x500")
label=Label(win1)

label1=Label(win1,text='帳號',bg='yellow',width=8,font='Helvetica 15 bold')
label2=Label(win1,text='密碼',bg='yellow',width=8,font='Helvetica 15 bold')
label1.grid(column=0,row=0)
label2.grid(column=0,row=1)

ent1=Entry(win1)
ent2=Entry(win1,show='*')
ent1.grid(column=1,row=0)
ent2.grid(column=1,row=1)

btn1=Button(win1,text='登入',width=15,command=login)
btn2=Button(win1,text='離開',width=15,command=win1.destroy)
btn1.grid(column=0,row=2)
btn2.grid(column=1,row=2)

win1.mainloop() #開啟視窗

```
